```python
# Paso 1: Cargar las librerías
# Librería para gráficas
import matplotlib.pyplot as plt

# Paso 2: Cargamos los datos
x = ['Hawaii', 'Florida', 'Missouri', 'Washington', 'California']
y = [3.6, 3.1, 3.0, 3.2, 3.4]

# Paso 3: Generamos la gráfica
plt.bar(x, y)
plt.title("Precio promedio de la gasolina por estado")
plt.xlabel("Estado")
plt.ylabel("Precio promedio")
plt.show()
```
