```python
# Paso 1: Cargar las librerías
# Librería para gráficas
import matplotlib.pyplot as plt

# Paso 2: Cargamos los datos
x = ['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo','Junio','Julio','Agosto','Septiembre','Octubre','Noviembre','Diciembre']
y = [42.6, 17.8, 82.2, 107.0, 115.4, 38.0, 86.0, 129.4, 32.2, 166.2, 65.0, 55.4]

# Generar la gráfica de Línea
plt.figure(figsize=(15, 5))
plt.plot(x, y, color="green", linestyle="-")
plt.title("Precipitaciones mensuales")
plt.xlabel("Meses")
plt.ylabel("Milimetros")
plt.grid()
plt.show()
```
