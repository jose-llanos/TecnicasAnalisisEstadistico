```python
# Paso 2: Importar las librerias
# Libreria para manejo de datos
import pandas as pd
# Libreria para prueba de Scheffé
import pingouin as pg

# Paso 3: Crear los datos
data = {
    'Grupo': ['A', 'A', 'A', 'B', 'B', 'B', 'C', 'C', 'C', 'D', 'D', 'D'],
    'Calificaciones': [85, 88, 86, 78, 82, 79, 92, 90, 91, 88, 86, 87]
}
df = pd.DataFrame(data)

# Paso 4: Calcular el ANOVA
pg.anova(data=df, dv='Calificaciones', between='Grupo')

# Paso 5: Prueba de Scheffé
pg.pairwise_tukey(data=df, dv='Calificaciones', between='Grupo')
```
